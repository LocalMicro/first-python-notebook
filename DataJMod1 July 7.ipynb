```python
# Chapter 16: Hello Cleaning.
# How to use 'apply' to make new column of double entries like "Sean Parker" and "Sean Parker & Affiliates"
```

```python
def combine_names(row):
    if row.contributor_fullname.startswith('SEAN PARKER'):
        return 'SEAN PARKER'
    return row.contributor_fullname
```

```python
# text says: You know (the above) is a function because it starts with def.
```

```python
# The apply command (see below) takes the function we created above (combine_names) and
# APPLIES it (what a surprise) to every row. 
# The 'axis = 1' means apply the combine_names function to every ROW.
# If we wanted to apply a function to every COLUMN, then we would use axis = 0.
```

```python
# In pandas this can be done for a DataFrame with the apply method. 
# To run the method row by row, pandas requires you pass the number 1 into the axis option.
```

```python
top_supporters.apply(combine_names, axis=1)
```

```python
# Notice that the two Sean Parker rows 
# now have the exact same label 'Sean Parker'
# instead of one having 'Sean Parker and Affiliated Entities'
```

```python
top_supporters['contributor_cleanname'] = top_supporters.apply(combine_names, axis=1)
```

```python
# We used the apply method and the 'contributor_cleanname' function to
# regroup the data then do sum totals again
```

```python
top_supporters.groupby(
    "contributor_cleanname"
).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
# above has now grouped the 2 Sean Parker contributions into one $8.75 million total
```
